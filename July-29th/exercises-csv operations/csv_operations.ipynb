```python
csv_data="""
Name,Math,Science,English
Amit,78,85,74
Priya,92,89,96
Kiran,65,70,60
Anita,88,95,90
Ravi,55,60,58"""
with open('students.csv','w')as f:
  f.write(csv_data)
```

```python
import pandas as pd
df=pd.read_csv('students.csv')
print(df.head(3))
```

```python
df['Average']=df[['Math','Science','English ']].mean(axis=1)
high=df.sort_values(by='Average',ascending=False).head(1)
print(high)
```

```python
m_80_science=df[df['Science']>80]
print(m_80_science)

avg_below_70=df[df['Average']<70]
print(avg_below_70)
```

```python
sort_english=df.sort_values(by='English ',ascending=False).head(3)
print(sort_english)
```

```python
import matplotlib.pyplot as plt
df.plot(kind='bar',x='Name',y=['Average'])
plt.grid(True)
plt.ylabel('Average score')
plt.show()
```

```python
df.plot(x='Name',y=['Math','Science','English '],marker='o',figsize=(8,5))
plt.title('Subject scores by student')
plt.xlabel('Student')
plt.ylabel('Score')
plt.grid(True)
plt.show()
```

```python
hight_subject=df[['Math','Science','English ']].sum()
highest_subject=hight_subject.idxmax()
print(f"Highest total score of student in a subject:{highest_subject}")

df.to_csv('students_summary.csv')
```
